## 1 . Import et analyse du jeu de donnÃ©es

Source : https://public.opendatasoft.com/explore/dataset/openaq/export/?disjunctive.city&disjunctive.location&disjunctive.measurements_parameter

```python
import pandas as pd
import numpy as np

df = pd.read_csv('openaq.csv', on_bad_lines='skip', delimiter=";")
#vÃ©rifications du type de donnÃ©es par colonne
print(df.dtypes)
df.head()
```

## 2 . Nettoyage du jeu de donnÃ©es

**Supprimer les donnÃ©es qui ne concernent pas la journÃ©e du 06-08-2021.**

```python
#conversion en utf-8 : cela force Ã  se dÃ©barasser des lignes dont les caractÃ¨res sont illisibles
for row in df :
  row.encode('utf-8','ignore').decode("utf-8")
  
#vÃ©rifications du type de donnÃ©es par colonne
print(df.dtypes)
df.head()
```

```python
df = df.loc[df["Last Updated"].str.contains("2021-08-06")]
df
```

**VÃ©rification qu'il n'y a pas de valeur de mesure manquantes**

```python
# affichage des valeurs manquantes
print("\nAffichage des valeurs nulles:")
display(df.loc[df['Value'].isnull()])
display(df.loc[df['Pollutant'].isnull()])
display(df.loc[df['City'].isnull()])
```

**VÃ©rification des valeurs nÃ©gatives**

```python
# suppression des valeurs nÃ©gatives et nulle
toDrop = df.loc[df['Value']<=0]
df = df.drop(toDrop.index)
```

**Pour chaque polluant, vÃ©rifier qu'on n'a qu'une seule mesure utilisÃ©e. Autrement dit, supprimer les donnÃ©es qui correspondent Ã  la mesure minoritaire ou moins cohÃ©rente.**


```python
# CrÃ©ation d'un dataframe qui ne contient que la valeur max pour chaque ville et chaque polluant
# on pourrait aussi conserver la mesure mediane avec la fonction .med()
PollutantCount = df.groupby(['Pollutant','City'])['Value'].max()

#jointure intÃ©rieure des deux dataframe : autrement dit, suppression des donnÃ©es correspondant Ã  la mesure minoritaire sauf s'il est est strictement Ã©gale au max
df = pd.merge(df, PollutantCount, on=['Pollutant', 'City', 'Value'], how='inner')

# Suppression des doublons basÃ©s sur la combinaison de colonnes 'Pollutant', 'City' et 'Value'
df.drop_duplicates(subset=['Pollutant', 'City', 'Value'], inplace=True)

df
```

**Regarder s'il y a des valeurs aberrantes**

DÃ©termination des quartiles : Q1 et Q3  
Quartile :Chacune des trois valeurs qui divisent les Ã©lÃ©ments d'une distribution statistique en quatre parties d'effectif Ã©gal.

```python
#crÃ©ations des quartiles
Q1 = round(df.groupby(['Pollutant'])['Value'].quantile(0.25),2)
Q2 = round(df.groupby(['Pollutant'])['Value'].std(),2)
Q3 = round(df.groupby(['Pollutant'])['Value'].quantile(0.75),2)

dfStats = pd.DataFrame({'Q1 value': Q1, 'Q2 value': Q2, 'Q3 value': Q3})
dfStats 

```

```python
#visualisation
dfInfQ1 = df.loc[df['Value'] < df['Pollutant'].map(Q1)].sort_values('Value')
dfSupQ3 = df.loc[df['Value'] > df['Pollutant'].map(Q3)].sort_values('Value', ascending = False)
display(dfInfQ1)
display(dfSupQ3)
```

```python
#crÃ©ation d'un dataframe Outliers
outliers = pd.concat([dfInfQ1, dfSupQ3]).sort_values('Value')
outliers
```

```python
#analyse du df final
ValuesPerCountry= df.value_counts(['Country Label'])
print(ValuesPerCountry)
```

```python
#export des 2 df sous format csv. Index = false Ã©vite de crÃ©er une colonne index pour les Ã©quipes compta qui y comprennent rien 
df.to_csv('air_quality_v2.csv', index=False)
outliers.to_csv('outlier.csv', index=False)
ValuesPerCountry.to_csv('countries.csv')
```
